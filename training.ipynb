```python
def get_unique_values(data: pd.DataFrame, column_name: str):
    """
    Get all unique values in a specified column of a DataFrame.

    Parameters:
    - data (pd.DataFrame): The dataset as a pandas DataFrame.
    - column_name (str): The name of the column to find unique values for.

    Returns:
    - list: A list of unique values in the column, or None if the column is not found.
    """
    if column_name in data.columns:
        unique_values = data[column_name].unique()
        print(f"Unique values in column '{column_name}':")
        return unique_values.tolist()  # Convert to a list for easy usage
    else:
        print(f"Column '{column_name}' not found in the dataset.")
        return None
```

```python
for feature in selected_features:
    unique_values = get_unique_values(data, feature)
    if unique_values is not None:
        print(f"Unique values in column '{feature}': {unique_values}")
        print(f"Number of unique values in column '{feature}': {len(unique_values)}")
        missing_values = data[feature].isnull().sum()
        print(f"Number of missing values in column '{feature}': {missing_values}")
        data = data.dropna(subset=selected_features, how='all')
    else:
        print(f"No unique values found for column '{feature}'")

print(get_unique_values(data, selected_features[1]))
```

```python
for index, row in data.iterrows():
    positions = row['player_positions'].split(',')
    if positions[0] in ['LW', 'RW', 'CF', 'ST']:
        data.at[index, 'player_positions'] = 'attacker'
    if positions[0] in ['LM', 'RM', 'CM', 'CAM', 'LAF', 'RAF']:
        data.at[index, 'player_positions'] = 'midfielder'
    if positions[0] in ['LB', 'LWB', 'LCB', 'CB', 'RCB', 'RWB', 'RB', 'CDM', 'LDM', 'RDM']:
        data.at[index, 'player_positions'] = 'defender'
    if positions[0] in ['GK']:
        data.at[index, 'player_positions'] = 'goalkeeper'
```

```python
df_selected_features = data[selected_features]
print(df_selected_features.shape)
```

```python
encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(df_selected_features[['player_positions']])
position_feature_names = encoder.get_feature_names_out(['player_positions'])
```

```python
numerical_features = ['pace','shooting' ,'passing' ,'dribbling','defending', 'physic']
scaler = MinMaxScaler()
numerical_scaled = scaler.fit_transform(data[numerical_features])
```

```python
combined_features = np.hstack([encoded, numerical_scaled])
feature_names = list(position_feature_names) + numerical_features
combined_df = pd.DataFrame(combined_features, columns=feature_names)
```

```python
similarity_matrix = cosine_similarity(combined_features)

# Display similarity matrix for player names
similarity_df = pd.DataFrame(similarity_matrix, index=data['short_name'], columns=data['short_name'])
```

```python
player_index = data[data['short_name'] == 'E. Džeko'].index[0]
similar_players = similarity_df.iloc[player_index].sort_values(ascending=False).index[1:3]
print(f"\nPlayers most similar to Player A: {list(similar_players)}")

```

```python
# 5. Process user input
def process_user_input(position, speed, physic, defense, dribbling, shooting, passing):
    # Encode position
    user_position_encoded = encoder.transform([[position]])
    # Encode body_type
    # Scale numerical attributes
    user_numerical = scaler.transform([[speed, physic, defense, dribbling, shooting, passing]])
    # Combine all user features
    user_features = np.hstack([user_position_encoded, user_numerical])
    return user_features

# 6. Find most similar players
def find_similar_players(user_features, combined_features, player_names, top_n=10):
    # Compute similarity
    similarities = cosine_similarity(user_features, combined_features).flatten()
    # Get top N most similar players
    
    
    similar_indices = np.argsort(-similarities)[:top_n]
    # print(similar_indices)
    similar_players_names = df_selected_features.iloc[similar_indices].values
    return similar_players_names
```

```python
# 85,92,91,95,34,65
user_input = {
    'player_positions': 'attacker',
    'speed': 99,
    'physic': 99,
    'defense': 99,
    'dribbling': 99,
    'shooting': 99,
    'passing': 99
}

# Process user input
user_features = process_user_input(
    user_input['player_positions'],
    user_input['speed'],
    user_input['physic'],
    user_input['defense'],
    user_input['dribbling'],
    user_input['shooting'],
    user_input['passing']
)
```

```python
similar_players = find_similar_players(user_features, combined_features, data['short_name'])

print(similar_players)
```
